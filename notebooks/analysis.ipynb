# Visualize the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/healthcare-dataset-stroke-data.csv')
print(df.head(10))
```

## Dataframe describe and info

```python
print(df.describe())
print(df.info())
```

## Target variable - stroke

```python
stroke_counts = df["stroke"].value_counts()
stroke_percentages = df["stroke"].value_counts(normalize=True) * 100
```

### Bar chart

```python
stroke_counts.plot(kind="bar", color=['skyblue', 'salmon'])
plt.title("Stroke vs No Stroke")
plt.xlabel("Stroke Status")
plt.ylabel("Number of Patients")
plt.xticks(ticks=[0, 1], labels=['No Stroke', 'Stroke'], rotation=0)
plt.show()
```

### Stroke Percentage

```python
stroke_percentages.plot(kind="bar", color=['skyblue', 'salmon'])
plt.title("Stroke Percentages")
plt.xlabel("Stroke")
plt.ylabel("Percentage of Patients")
plt.xticks(ticks=[0, 1], labels=['No Stroke', 'Stroke'], rotation=0)
plt.show()
```

```python
def load_data(filepath: str) -> pd.DataFrame:
    """Load CSV data into a pandas DataFrame."""
    return pd.read_csv(filepath)

data = load_data('../data/healthcare-dataset-stroke-data.csv')
# print(data.head(10))

def fill_missing_values_median(data: pd.DataFrame) -> pd.DataFrame:
    """Fill missing BMI values using the median."""
    return data.assign(bmi=data['bmi'].fillna(data['bmi'].median()))

filled_data = fill_missing_values_median(data)
# print(filled_data.head(10))

def encode_categorical_features(data: pd.DataFrame) -> pd.DataFrame:
    """Convert categorical columns to numerical format (one-hot encoding)."""
    return pd.get_dummies(data,
                          columns=['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status'],
                          drop_first=True)

encoded_data = encode_categorical_features(filled_data)
# print(encoded_data.to_string())

from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

def scale_numerical_features(data: pd.DataFrame) -> pd.DataFrame:
    """Standardize numerical features."""
    numerical_columns = ['age', 'avg_glucose_level', 'bmi']
    copy_data = data.copy()

    scaled = scale.fit_transform(copy_data[numerical_columns])
    copy_data[numerical_columns] = scaled
    return copy_data

scaled_data = scale_numerical_features(encoded_data)
# print(type(scaled_data))
# print(scaled_data)

def split_features_and_target(data: pd.DataFrame, target_column: str = 'stroke'):
    """Split dataset into features (X) and target (y)."""
    X = data.drop(columns=[target_column])
    y = data[target_column]
    return X, y

X, y = split_features_and_target(scaled_data)

print(X.shape)
print(y.value_counts())
```
